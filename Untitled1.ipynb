```python
# ======================
# Import Libraries
# ======================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier
from imblearn.over_sampling import SMOTE
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# ======================
# Configuration
# ======================
pd.set_option('display.max_columns', 50)
plt.style.use('ggplot')

# ======================
# Helper Functions
# ======================
def load_data(file_path: str) -> pd.DataFrame:
    """Load and preprocess raw data."""
    df = pd.read_excel(file_path, sheet_name="Form Responses 1")
    
    # Select and rename columns
    df = df.rename(columns={
        'highest level of education completed': 'education',
        'job title': 'job_title',
        'overall years of professional experience': 'experience'
    })[['industry', 'job_title', 'annual salary', 'country', 'education', 'experience']]
    
    return df

def preprocess_data(df: pd.DataFrame) -> pd.DataFrame:
    """Clean and transform data."""
    # Salary categorization
    df['salary_category'] = pd.cut(
        df['annual salary'],
        bins=[0, 50_000, 100_000, float('inf')],
        labels=['low', 'medium', 'high']
    )
    
    # Clean job titles
    df['job_title'] = (
        df['job_title']
        .str.replace(r'\b(I|II|III|Senior|Junior|Lead|Manager|Director)\b', '', regex=True)
        .str.strip()
        .str.lower()
    )
    
    # Bin experience
    df['experience_binned'] = pd.cut(
        df['experience'].str.extract(r'(\d+)')[0].astype(float),
        bins=[0, 5, 10, 20, 50],
        labels=['0-5', '5-10', '10-20', '20+']
    )
    
    # Drop missing values
    return df.dropna()

# ======================
# Main Pipeline
# ======================
def main():
    # Load and preprocess data
    df = load_data("Ask A Manager Salary Survey 2021 (Responses).xlsx")
    df = preprocess_data(df)
    
    # ======================
    # Feature Engineering
    # ======================
    # Define preprocessing steps
    preprocessor = ColumnTransformer(
        transformers=[
            ('cat', OneHotEncoder(handle_unknown='ignore'), 
             ['industry', 'country', 'education', 'job_title', 'experience_binned'])
        ],
        remainder='drop'
    )
    
    # Split data
    X = df.drop(columns=['annual salary', 'salary_category'])
    y = df['salary_category']
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    # ======================
    # Handle Class Imbalance
    # ======================
    smote = SMOTE(random_state=42)
    X_resampled, y_resampled = smote.fit_resample(preprocessor.fit_transform(X_train), y_train)
    
    # ======================
    # Model Training
    # ======================
    model = Pipeline([
        ('preprocessor', preprocessor),
        ('classifier', RandomForestClassifier(
            n_estimators=200,
            class_weight='balanced',
            random_state=42
        ))
    ])
    
    model.fit(X_train, y_train)
    
    # ======================
    # Evaluation
    # ======================
    y_pred = model.predict(X_test)
    
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
    print("\nClassification Report:\n", classification_report(y_test, y_pred))
    
    # ======================
    # Visualizations
    # ======================
    # Class distribution
    plt.figure(figsize=(10, 5))
    sns.countplot(data=df, x='salary_category', hue='salary_category', palette='viridis', legend=False)
    plt.title("Salary Category Distribution", fontsize=14)
    plt.xlabel("Salary Category", fontsize=12)
    plt.ylabel("Count", fontsize=12)
    plt.show()
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=model.classes_, 
                yticklabels=model.classes_)
    plt.title("Confusion Matrix", fontsize=14)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
    
    # Feature importance
    feature_names = model.named_steps['preprocessor'].transformers_[0][1].get_feature_names_out(
        ['industry', 'country', 'education', 'job_title', 'experience_binned']
    )
    
    importances = model.named_steps['classifier'].feature_importances_
    top_features = pd.DataFrame({
        'feature': feature_names,
        'importance': importances
    }).nlargest(15, 'importance')
    
    plt.figure(figsize=(12, 6))
    sns.barplot(data=top_features, x='importance', y='feature', palette='rocket')
    plt.title("Top 15 Important Features", fontsize=14)
    plt.xlabel("Importance Score", fontsize=12)
    plt.ylabel("")
    plt.show()

# ======================
# Run Pipeline
# ======================
if __name__ == "__main__":
    main()
```
