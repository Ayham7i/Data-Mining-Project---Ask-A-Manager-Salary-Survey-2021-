# K-Mean Algorithm

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Load the dataset
df = pd.read_excel("Ask A Manager Salary Survey 2021 (Responses).xlsx")

# Step 1: Filter USD entries
df = df[df["currency"] == "usd"]

# Step 2: Convert experience ranges to numerical values
experience_mapping = {
    "1 year or less": 0.5,
    "2 - 4 years": 3,
    "5-7 years": 6,
    "8 - 10 years": 9,
    "11 - 20 years": 15.5,
    "21 - 30 years": 25.5,
    "31 - 40 years": 35.5,
    "41 years or more": 45
}
df["experience"] = df["overall years of professional experience"].map(experience_mapping)

# Step 3: Encode education levels
education_mapping = {
    "high school": 1,
    "some college": 2,
    "college degree": 3,
    "master's degree": 4,
    "phd": 5
}
df["education"] = df["highest level of education completed"].map(education_mapping)

# Step 4: Handle missing/other monetary compensation
df["other monetary comp"] = df["other monetary comp"].fillna(0)

# Step 5: Create total compensation
df["total_compensation"] = df["annual salary"] + df["other monetary comp"]

# Step 6: Drop rows with missing values in the selected features
df = df.dropna(subset=["total_compensation", "experience", "education"])

# Step 7: Verify no missing values remain
print("Missing values after preprocessing:")
print(df[["total_compensation", "experience", "education"]].isnull().sum())

# Step 8: Select features for clustering
features = df[["total_compensation", "experience", "education"]]

# Step 9: Standardize features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Step 10: Verify no NaN values in scaled_features
if np.isnan(scaled_features).any():
    raise ValueError("scaled_features contains NaN values. Check preprocessing steps.")

# Step 11: Determine optimal K using the Elbow Method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(8, 5))
plt.plot(range(1, 11), inertia, marker="o", linestyle="--")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal K")
plt.show()

# Step 12: Apply K-means with K=4
kmeans = KMeans(n_clusters=4, random_state=42)
df["cluster"] = kmeans.fit_predict(scaled_features)

# Step 13: Visualize clusters in 2D
plt.figure(figsize=(10, 6))
for cluster in range(4):
    plt.scatter(
        df[df["cluster"] == cluster]["total_compensation"],
        df[df["cluster"] == cluster]["experience"],
        label=f"Cluster {cluster}"
    )
plt.xlabel("Total Compensation")
plt.ylabel("Experience (Years)")
plt.title("2D Clustering: Total Compensation vs Experience")
plt.legend()
plt.show()

# Step 14: Visualize clusters in 3D
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")
for cluster in range(4):
    ax.scatter(
        df[df["cluster"] == cluster]["total_compensation"],
        df[df["cluster"] == cluster]["experience"],
        df[df["cluster"] == cluster]["education"],
        label=f"Cluster {cluster}"
    )
ax.set_xlabel("Total Compensation")
ax.set_ylabel("Experience (Years)")
ax.set_zlabel("Education Level")
ax.set_title("3D Clustering: Total Compensation, Experience, and Education")
plt.legend()
plt.show()

# Step 15: Analyze clusters
cluster_summary = df.groupby("cluster").agg({
    "total_compensation": ["mean", "median"],
    "experience": ["mean", "median"],
    "education": ["mean", "median"]
})
print(cluster_summary)
```
